```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
def JonesErr(filename):
    J = np.loadtxt(filename,delimiter='\t',dtype=str)
    Targets = []
    for i in range(1,200):
        nums = ''.join(e for e in J[i,0] if e!='{' and e!="\"" and e!= "}").split(',')
        for j in range(len(nums)):
            nums[j] = float(nums[j])
        Targets.append(nums)
    # print(Targets)
    Results = []
    bads = []
    for i in range(1,200):
        nums = ''.join(e for e in J25[i,2][:J25[i,2].find(", {{")] if e!='{' and e!="\"" and e!= "}").split(',')
        for j in range(len(nums)):
            if nums[j] == " -Sign[Eigenvalues[]]":
                nums[j] = np.nan
                bads.append(i-1)
                print('row',i,'!!!',nums)
            nums[j] = float(nums[j])
        Results.append(nums)
    # print(Results)
    for i in range(len(bads)):
        del Targets[bads[i]-i]
        del Results[bads[i]-i]
    Results_npy = np.array([np.array(r) for r in Results])
    Targets_npy = np.array([np.array(t) for t in Targets])
    L2_err = np.linalg.norm(Targets_npy - Results_npy,2,0)
    ax = plt.subplot(2,2,1)
    ax.scatter(Targets_npy[:,0],Results_npy[:,0])
    ax.plot([0,25],[0,25],c='r')
    ax = plt.subplot(2,2,2)
    ax.scatter(Targets_npy[:,1],Results_npy[:,1])
    ax.plot([0,20],[0,20],c='r')
    ax = plt.subplot(2,2,3)
    ax.scatter(Targets_npy[:,2],Results_npy[:,2])
    ax.plot([-10,10],[-10,10],c='r')
    ax = plt.subplot(2,2,4)
    ax.scatter(Targets_npy[:,3],Results_npy[:,3])
    plt.show()
    return L2_err
```
